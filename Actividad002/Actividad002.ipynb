#### Voy a dividir la data para implementar posteriormente un modelo

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['concrete_comprenssive_strength'])
Y = df['concrete_comprenssive_strength']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=30, random_state=150)
```

#### Estandarizo las variables que no lo están.

```python
from sklearn.preprocessing import StandardScaler

X_train_std = StandardScaler().fit_transform(X_train)
X_test_std = StandardScaler().fit_transform(X_test)
```

## Implemento el modelo de LinearGAM

```python
from pygam import LinearGAM

model_gam = LinearGAM().fit(X_train_std, y_train) 
model_gam_pred = model_gam.predict(X_test_std)
model_gam_pred[:10]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae_gam = mean_absolute_error(y_test, model_gam_pred)
mse_gam = mean_squared_error(y_test, model_gam_pred)
```

```python
print("Hiperparametro lambda: {0}".format(model_gam.lam))
print("MAE: ", mae_gam)
print("MSE: ", mse_gam)
```

### Modelo LinearGAM con GridSearch

```python
search_params = {'lam':np.logspace(-3, 3, 11)}
model_gam2 = LinearGAM().gridsearch(X_train_std, y_train, **search_params)
```

```python
model_gam2_pred = model_gam2.predict(X_test_std)
```

```python
mae_gam2 = mean_absolute_error(y_test, model_gam2_pred)
mse_gam2 = mean_squared_error(y_test, model_gam2_pred)
```

```python
print("Hiperparamentro lambda: ".format(model_gam2.lam))
print("MAE: ", mae_gam2)
print("MSE: ", mse_gam2)
```

### Dependencia Parcial

Tomaré como mejor modelo el `Model GAM 2` que se aplicó con gridsearch.

```python
from pygam.utils import generate_X_grid
```

```python
x_grid = generate_X_grid(model_gam2)
plt.figure(figsize=(10,6))
attribute = X_train.columns
cols = 4; rows = int(len(attribute) / cols)

for index, value in enumerate(range(len(attribute))):
    plt.subplot(rows, cols, index + 1)
    # extraemos la dependencia parcial y sus intervalos de confianza al 95%
    partial_dep, confidence_intervals = model_gam2.partial_dependence(x_grid, feature=index + 1, width=.95)
    plt.plot(x_grid[:, value], partial_dep, color='dodgerblue')
    
    # Generamos una visualización de los intervalos de confianza
    plt.fill_between(x_grid[:, value],
                     confidence_intervals[0][:, 0],
                     confidence_intervals[0][:, 1], 
                     color='tomato', alpha=0.5)
    #Nombre del atributo
    plt.title(attribute[value])
    plt.plot(X_train[attribute[value]], 
             [plt.ylim()[0]] * len(X_train[attribute[value]]),
             '|', color='orange', alpha=.5)
plt.tight_layout();
```
