## Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, roc_auc_score

from keras.models import Sequential
from keras.layers import Dense

plt.style.use("seaborn") # Gráficos estilo seaborn
```

```python
df = pd.read_csv('data/dataset_bc.csv')
df.head()
```

```python
df = df.drop(columns=['id','Unnamed: 32'], axis=1)
df.head()
```

```python
df.isnull().any()
```

```python
df['diagnosis'] = df['diagnosis'].replace(['M', 'B'], [1, 0])
df['diagnosis'].value_counts()
```

```python
df.describe()
```

```python
df.shape
```

```python
plt.figure(figsize=(20,20))

for index, value in enumerate(df):
    plt.subplot(6,6,index+1)
    plt.hist(df[value], alpha=.5, color="grey")
    plt.axvline(df[value].mean(), ls="--", lw=2, color="tomato", label="Media")
    plt.xlabel(value)
    plt.legend()
```

```python
plt.figure(figsize=(22,20))

for index, value in enumerate(df):
    plt.subplot(6, 6, index + 1)
    sns.distplot(df[value])
    plt.xlabel(value)
```

```python
corr_df = df.corr()
plt.figure(figsize=(18,16))
sns.heatmap(corr_df, cmap="Blues", annot=True);
```

```python
def plot_distribution(data_select):
    plt.figure(figsize=(20,10))
    s = sns.FacetGrid(df, hue='diagnosis', aspect=2.5, palette={0:'skyblue', 1:'red'})
    s.map(sns.kdeplot, data_select, shade=True, alpha=0.5)
    s.add_legend()
    s.set_axis_labels(data_select, 'Proportion')
    s.fig.suptitle(data_select)
    

plot_distribution('radius_mean');
```

```python
plot_distribution('texture_mean');
plot_distribution('perimeter_mean')
```

# Divido la data para implementar los modelos

```python
X,Y = df.drop(columns=['diagnosis']), df['diagnosis']
```

### Hago over sampling porque la data se encuentra sesgada

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_resampled, Y_resampled = ros.fit_sample(X,Y)
```

```python
X_resampled = pd.DataFrame(X_resampled)
Y_resampled = pd.DataFrame(Y_resampled)
```

### Ahora divido la data en mi matriz de entrenamiento y validación, ya una vez balanceada la data.

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled,
                                                   Y_resampled,
                                                   test_size=0.33,
                                                   random_state=14785)
```

```python
X_train_std = StandardScaler().fit_transform(X_train)
X_test_std = StandardScaler().fit_transform(X_test)
```

### Tengo la data ya dividida en entrenamiento y validación, y estandarizada para empezar a implementar los modelos.

Los siguientes modelos serán entrenados:
    1. Logistic Regression with Scikit-Learn
    2. Gradient boosting with XGBoost
    3. Neural Network with Keras

### 1.- Logistic Regression

```python
model_lg = LogisticRegression().fit(X_train_std, y_train)
lg_pred = model_lg.predict(X_test_std)
lg_pred [:20]
```

### 2.- XGBoost

```python
model_xgb = XGBClassifier().fit(X_train_std, y_train)
xgb_pred = model_xgb.predict(X_test_std)
xgb_pred[:20]
```

### 3.- Keras Network

```python
model_keras = Sequential()
model_keras.add(Dense(1028, input_dim=30, activation="relu"))
model_keras.add(Dense(31, activation='relu'))
model_keras.add(Dense(1, activation='sigmoid'))
```

```python
# compile model
model_keras.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model_keras.fit(X_train_std, y_train, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model_keras.evaluate(X_train_std, y_train)
print("\n%s: %.2f%%" % (model_keras.metrics_names[1], scores[1]*100))
```

# Evaluo los modelos

### 1.- Logistic Regression

#### Matriz Confusion

```python
lg_confusion = confusion_matrix(y_test, lg_pred)
lg_confusion
```

#### Accuracy

```python
accuracy_lg = accuracy_score(y_test, lg_pred)
print("Accuracy Logistic Regression: %.2f%%" %(accuracy_lg * 100))
```

#### Con este dato solo obtengo el porcentaje de exactitud de los 0, es decir, cuando no es Maligno el cancer, implementaré el `classification_report` para obtener el porcentaje de exactitud para `0 y 1`.

```python
print(classification_report(y_test,lg_pred))
```

#### Un buen modelo de clasificación también debe considerar cuál es el rango de errores en el modelo con el que se trabaja. Para ello utilizamos la curva ROC (Receiving Operating Characteristics), que evalúa la relación entre ambos errores condicional en todo el rango del clasificador.

```python
# reestimo los valores del modelo para obtener la probabilidad
lg_pred_proba = model_lg.predict_proba(X_test_std)[:,1]

# generamos los objetos de roc_cruve
false_positive, true_positive, threshold = roc_curve(y_test, lg_pred_proba)
```

```python
# Plot ROC curve
plt.title('Curva ROC')
plt.plot(false_positive, true_positive, lw=1)
plt.plot([0, 1], ls="--", lw=1)
plt.plot([0, 0], [1, 0] , c='limegreen', lw=3), plt.plot([1, 1] , c='limegreen', lw=3)
plt.ylabel('Falsos Positivos')
plt.xlabel('Verdaderos Positivos');
```

```python
accuracy_roc = roc_auc_score(y_test, lg_pred_proba)
print("Área bajo la curva ROC: %.2f%%" %(accuracy_roc * 100))
```

#### Haré que mi modelo de LogisticRegression sea aún más robusto, de tal manera implementaré una validación cruzada.

```python
model_lg_cv = cross_val_score(LogisticRegression(),
                             X_train_std,
                             y_train,
                             cv=10,
                             scoring="f1")
model_lg_cv
```

```python
for i, n in enumerate(model_lg_cv):
    print("El puntaje F1 para la muestra {0} es {1}".format(i+1, round(n, 2)))
```

```python
accuracy_lg_cruzado = round(np.mean(model_lg_cv),2)
print("El puntaje promedio F1 del modelo es:", accuracy_lg_cruzado)
print("La desviación estandar del modelo es de:", round(np.std(model_lg_cv),2))
```

### 2.- XGBoost

```python
eval_set = [(X_test_std, y_test)]
model_xgb = XGBClassifier().fit(X_train_std, y_train, early_stopping_rounds=10, eval_metric="logloss",
                               eval_set=eval_set, verbose=True)
```

```python
from xgboost import plot_importance
# plot feature importance
plot_importance(model_xgb)
plt.show()
```

```python
xgb_pred = model_xgb.predict(X_test_std)
xgb_pred[:20]
```

```python
accuracy_xgb = accuracy_score(y_test, xgb_pred)
print("Accuracy XGB: %.2f%%" % (accuracy_xgb * 100))
```

#### Haré una búsqueda de hiperparametros para mejorar el modelo, donde emplearé parametros como `learning_rate`, `n_estimators`, `max_depth`, utilizando una validación cruzada.

```python
# grid search
n_estimators = [50, 100, 150, 200]
max_depth = [2, 4, 6, 8]
learning_rate = [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth, learning_rate=learning_rate)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
grid_search = GridSearchCV(XGBClassifier(), param_grid, scoring="neg_log_loss", n_jobs=-1, cv=kfold, verbose=1)
result = grid_search.fit(X_train_std, y_train)
```

```python
print("Best: %f using: %s" % (result.best_score_, result.best_params_))
```

#### Obtengo la mejor combinanción de parametros encontrados, ahora procedo a implementar mi modelo con esta.

```python
model_xgb = XGBClassifier(n_estimators=200, max_depth=4, learning_rate=0.3, subsample=0.75, colsample_bytree=1)
```

```python
model_xgb.fit(X_train_std, y_train)
```

```python
xgb_pred = model_xgb.predict(X_test_std)
xgb_pred[:20]
```

#### Ahora calculo el Accuracy del Modelo.

```python
accuracy_xgb_cruzado = accuracy_score(y_test, xgb_pred)
print("Accuracy XGB: %.2f%%" %(accuracy_xgb_cruzado * 100))
```

### 3.- Keras

```python
# evaluate the model
scores = model_keras.evaluate(X_train_std, y_train)
print("\n%s: %.2f%%" % (model_keras.metrics_names[1], scores[1]*100))
```

```python
keras_pred = model_keras.predict(X_test_std)
keras_pred[:20]
```
